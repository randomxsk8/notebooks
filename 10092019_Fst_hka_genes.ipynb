```python
def fst_func(chrom, start, end):
    
    pos_sel = sort_pos[chrom].intersect_range(start, end)
    
    ac1 = ac_m[chrom].take(pos_sel)
    ac2 = ac_s[chrom].take(pos_sel)
    
    num, den = allel.hudson_fst(ac1, ac2)
    fst = np.sum(num) / np.sum(den)
    
    return fst
```
