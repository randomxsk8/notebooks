```python
hka_coluzzii_3L = hka_coluzzii.loc[(hka_coluzzii['chrom'] == '3L')]
hka_coluzzii_3L
```

```python
hka_coluzzii_3R = hka_coluzzii.loc[(hka_coluzzii['chrom'] == '3R')]
hka_coluzzii_3R
```

```python
def center_pos(r):
    start = r.start
    end = r.end
    center = ((end - start) / 2) + start
    return int(center)
```

```python
tbl_genes  = (etl
    .fromgff3('data/Anopheles-gambiae-PEST_BASEFEATURES_AgamP4.2.sorted.gff3.gz')
    .unpackdict('attributes')
    .cut(2,3,4,9)
    .select(lambda rec: rec.type == 'gene')
    .rename('ID', 'gene')
    .addfield('center_gene', center_pos)
    .todataframe()
)
tbl_genes.head()
```

```python
tbl_pos_3L = pd.merge(hka_coluzzii_3L, tbl_genes, on='gene', how='inner').drop(columns=['type','start','end'])
tbl_pos_3L.to_csv('tables/tbl_pos_3L.tab')
tbl_pos_3L
```

```python
tbl_pos_3L = pd.merge(hka_coluzzii_3L, tbl_genes, on='gene', how='inner').drop(columns=['type','center_gene'])
tbl_pos_3L.to_csv('tables/tbl_pos_3L.tab')
tbl_pos_3L
```
